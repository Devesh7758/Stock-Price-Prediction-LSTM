## Step 2: Import the Libraries
We import essential libraries like `yfinance`, `pandas`, `matplotlib`, `statsmodels` for ARIMA, and `TensorFlow` for LSTM.


```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

## Step 3: Download Historical Stock Price Data
We download Apple Inc. stock prices from Yahoo Finance between 2015 and 2024 and plot the closing price.


```python
data = yf.download('AAPL', start='2015-01-01', end='2024-12-31')
data = data[['Close']]  # Keep only the 'Close' price
data.plot(title='Apple Stock Price (2015–2024)', figsize=(10, 4))
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid()
plt.show()
```

## Step 4: Forecasting with ARIMA Model
We use the ARIMA model (AutoRegressive Integrated Moving Average) to forecast the next 30 days of stock prices.


```python
from statsmodels.tsa.stattools import adfuller

# Step 1: Check if the data is stationary
result = adfuller(data['Close'])
print("ADF Statistic:", result[0])
print("p-value:", result[1])

# Step 2: If p-value > 0.05, do differencing to make it stationary
data_diff = data['Close'].diff().dropna()

# Step 3: Fit ARIMA model (p=5, d=1, q=0)
model = ARIMA(data['Close'], order=(5, 1, 0))
model_fit = model.fit()
print(model_fit.summary())

# Step 4: Forecast next 30 days
forecast = model_fit.forecast(steps=30)

# Step 5: Plot the results
plt.figure(figsize=(10, 4))
plt.plot(data.index, data['Close'], label='Historical Data')
future_dates = pd.date_range(start=data.index[-1], periods=31, freq='D')[1:]
plt.plot(future_dates, forecast, label='ARIMA Forecast', color='red')
plt.title('ARIMA Forecast for Next 30 Days')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid()
plt.show()
```

## Step 5: Prepare Data for LSTM Model
LSTM requires the data to be scaled and structured as sequences. We normalize the prices and create training sequences.


```python
# Step 5: Use full OHLCV data instead of only 'Close'
df = yf.download('AAPL', start='2015-01-01', end='2024-12-31')
df = df[['Open', 'High', 'Low', 'Close', 'Volume']]

# Normalize all 5 features
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

X = []
y = []
window = 60  # sequence length (past 60 days)

# Create training sequences with all features
for i in range(window, len(scaled_data)):
    X.append(scaled_data[i-window:i])
    y.append(scaled_data[i, 3])  # 3 = Close price (still the target)

X = np.array(X)
y = np.array(y)

print("Shape of X:", X.shape)  # should be (samples, 60, 5)
print("Shape of y:", y.shape)
```

## 🔧 Step 6: Build and Train the LSTM Model

We define a deep learning model using LSTM (Long Short-Term Memory) layers. This model takes in 60-day sequences of OHLCV data and learns to predict the next day's closing price. We use a dropout layer to reduce overfitting and train the model using the Adam optimizer and Mean Squared Error loss.


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Step 6: Build and train the LSTM model with 5-feature input
model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(X.shape[1], X.shape[2])))
model.add(LSTM(64))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()

# Train the model
model.fit(X, y, epochs=20, batch_size=32)
```

## 🔮 Step 7: Make Predictions on Test Data

We now prepare the last 60 days of data as test input and use the trained LSTM model to forecast the next 30 days of closing prices. The predicted values are then scaled back to original price values using inverse transformation.


```python
# Use last 60 days from original scaled data to make prediction
last_60_days = scaled_data[-60:]
last_60_days = np.reshape(last_60_days, (1, 60, 5))  # Reshape for model

# Predict the next day's close price
predicted_price_scaled = model.predict(last_60_days)
predicted_price = scaler.inverse_transform(
    np.concatenate((np.zeros((1, 3)), predicted_price_scaled, np.zeros((1, 1))), axis=1)
)[0][3]  # Extract predicted 'Close'

print("Predicted next day's closing price:", predicted_price)
```

## 📈 Step 8: Plot Actual vs Predicted Prices

To visually evaluate the model's performance, we compare the actual closing prices from the last 30 days with the predicted prices from our model. This helps us understand how closely the model's forecast aligns with reality.


```python
# Split data into training and testing sets (last 100 days as test)
split = -100
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Predict on test set
predictions_scaled = model.predict(X_test)

# Pad predictions into full OHLCV shape to inverse transform properly
def rescale_predictions(preds, original):
    padded = np.zeros((preds.shape[0], 5))
    padded[:, 3] = preds[:, 0]  # Put predicted Close price in index 3
    return scaler.inverse_transform(padded)[:, 3]

predicted_prices = rescale_predictions(predictions_scaled, scaled_data)
actual_prices = scaler.inverse_transform(np.pad(
    y_test.reshape(-1,1), ((0,0),(0,4)), mode='constant'))[:, 0]

# Plot actual vs predicted
plt.figure(figsize=(10,5))
plt.plot(actual_prices, label='Actual Price', color='black')
plt.plot(predicted_prices, label='Predicted Price', color='red')
plt.title("Actual vs Predicted Closing Prices (Test Data)")
plt.xlabel("Days")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

## 📉 Step 9: Evaluate the Model (RMSE and MAE)

To quantify the model's accuracy, we calculate the Root Mean Squared Error (RMSE) and Mean Absolute Error (MAE) between the actual and predicted prices. Lower values indicate better performance.


```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

rmse = np.sqrt(mean_squared_error(actual_prices, predicted_prices))
mae = mean_absolute_error(actual_prices, predicted_prices)

print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```
